### b. Handling Categorical Features

```python
X = df[['num_passengers', 'sales_channel', 'trip_type', 'purchase_lead',
       'length_of_stay', 'flight_hour', 'flight_day', 'route',
       'booking_origin', 'wants_extra_baggage', 'wants_preferred_seat',
       'wants_in_flight_meals', 'flight_duration']]
y = df['booking_complete']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42,
                                                     test_size = 0.2)
```

```python
X_train_num = X_train[['num_passengers', 'purchase_lead',
       'length_of_stay', 'flight_hour', 'wants_extra_baggage', 'wants_preferred_seat',
       'wants_in_flight_meals', 'flight_duration']]
X_test_num = X_test[['num_passengers', 'purchase_lead',
       'length_of_stay', 'flight_hour', 'wants_extra_baggage', 'wants_preferred_seat',
       'wants_in_flight_meals', 'flight_duration']]
```

```python
X_train_num
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse = False, dtype = np.int32)
X_train_ohe = ohe.fit_transform(X_train[['sales_channel', 'trip_type', 'flight_day']])
X_test_ohe = ohe.transform(X_test[['sales_channel', 'trip_type', 'flight_day']])

X_train_ohe = pd.DataFrame(X_train_ohe)
X_test_ohe = pd.DataFrame(X_test_ohe)
```

```python
X_train_ohe.columns = ['sale_channel_internet', 'sales_channel_mobile',
                       'trip_type_Circle', 'trip_type_OneWay', 'trip_type_RoundTrip',
                       'flight_day_Fri', 'flight_day_Mon', 'flight_day_Sat', 'flight_day_Sun',
                       'flight_day_Thu', 'flight_day_Tue', 'flight_day_Wed']
X_test_ohe.columns = ['sale_channel_internet', 'sales_channel_mobile',
                       'trip_type_Circle', 'trip_type_OneWay', 'trip_type_RoundTrip',
                       'flight_day_Fri', 'flight_day_Mon', 'flight_day_Sat', 'flight_day_Sun',
                       'flight_day_Thu', 'flight_day_Tue', 'flight_day_Wed']
```

```python
counts1 = df['route'].value_counts()
thershold = 500
repl1 = counts1[counts1 <= thershold].index
X_train_route = pd.get_dummies(X_train['route'].replace(repl1, 'uncommon'))
X_test_route = pd.get_dummies(X_test['route'].replace(repl1, 'uncommon'))
X_train_route.columns
```

```python
counts_origin = df['booking_origin'].value_counts()
thershold = 1000
repl = counts_origin[counts_origin <= thershold].index
X_train_origin = pd.get_dummies(X_train['booking_origin'].replace(repl, 'uncommon'))
X_test_origin = pd.get_dummies(X_test['booking_origin'].replace(repl, 'uncommon'))
X_train_origin.columns
```

```python
X_train_new = np.concatenate((X_train_num, X_train_ohe, X_train_route, X_train_origin),
                             axis = 1)
X_test_new = np.concatenate((X_test_num, X_test_ohe, X_test_route, X_test_origin), axis = 1)
```

```python
X_train_new = pd.DataFrame(X_train_new)
X_test_new = pd.DataFrame(X_test_new)
```

```python
X_train_new.columns = ['num_passengers', 'purchase_lead',
       'length_of_stay', 'flight_hour', 'wants_extra_baggage', 'wants_preferred_seat',
       'wants_in_flight_meals', 'flight_duration', 'sale_channel_internet', 'sales_channel_mobile',
        'trip_type_Circle', 'trip_type_OneWay', 'trip_type_RoundTrip',
        'flight_day_Fri', 'flight_day_Mon', 'flight_day_Sat', 'flight_day_Sun',
        'flight_day_Thu', 'flight_day_Tue', 'flight_day_Wed', 'AKLKUL', 'COKSYD', 'DMKKIX', 'DMKOOL', 'DMKPER', 'DMKSYD', 'DPSICN',
       'ICNSIN', 'ICNSYD', 'MELPEN', 'MELSGN', 'MELTPE', 'PENTPE', 'SGNSYD',
       'uncommon', 'Australia', 'China', 'India', 'Indonesia', 'Japan', 'Malaysia',
       'New Zealand', 'Singapore', 'South Korea', 'Taiwan', 'Thailand',
       'uncommon']
X_test_new.columns = ['num_passengers', 'purchase_lead',
       'length_of_stay', 'flight_hour', 'wants_extra_baggage', 'wants_preferred_seat',
       'wants_in_flight_meals', 'flight_duration', 'sale_channel_internet', 'sales_channel_mobile',
        'trip_type_Circle', 'trip_type_OneWay', 'trip_type_RoundTrip',
        'flight_day_Fri', 'flight_day_Mon', 'flight_day_Sat', 'flight_day_Sun',
        'flight_day_Thu', 'flight_day_Tue', 'flight_day_Wed', 'AKLKUL', 'COKSYD', 'DMKKIX', 'DMKOOL', 'DMKPER', 'DMKSYD', 'DPSICN',
       'ICNSIN', 'ICNSYD', 'MELPEN', 'MELSGN', 'MELTPE', 'PENTPE', 'SGNSYD',
       'uncommon', 'Australia', 'China', 'India', 'Indonesia', 'Japan', 'Malaysia',
       'New Zealand', 'Singapore', 'South Korea', 'Taiwan', 'Thailand',
       'uncommon']
```

# Gradient Boosting Classifier

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.decomposition import PCA
gbc = GradientBoostingClassifier(max_depth = 3, learning_rate = 0.1, random_state = 0)
gbc.fit(X_train_new, y_train)
print('Training Accuracy ', gbc.score(X_train_new, y_train))
print('Testing Accuracy ', gbc.score(X_test_new, y_test))
```

```python
# Feature Importance
importances = gbc.feature_importances_

# Normalize the importances to sum to 100%
importances = 100.0 * (importances / importances.sum())

# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]

# Get feature names
feature_names = X_train_new.columns if isinstance(X_train_new, pd.DataFrame) else np.arange(X_train_new.shape[1])

# Ensure the number of features to display does not exceed the actual number of features
num_features_to_display = min(10, len(importances))

# Print shapes for debugging
print(f"Shape of X_train_new: {X_train_new.shape}")
print(f"Length of importances: {len(importances)}")
print(f"Length of feature names: {len(feature_names)}")

# Print the feature ranking
print("Feature ranking:")
for i in range(num_features_to_display):
    print(f"{i + 1}. feature {feature_names[indices[i]]} ({importances[indices[i]]:.2f}%)")

# Plot feature importances
plt.figure(figsize=(10, 6))
plt.title("Top 10 Feature Importances")
plt.bar(range(num_features_to_display), importances[indices[:num_features_to_display]], align='center')
plt.xticks(range(num_features_to_display), feature_names[indices[:num_features_to_display]], rotation=90)
plt.xlim([-1, num_features_to_display])
plt.xlabel('Feature')
plt.ylabel('Importance Score (%)')
plt.show()
```

```python

print(f"Top {num_features_to_display} feature names: {feature_names[indices[:num_features_to_display]]}")
print(f"Top {num_features_to_display} importances: {importances[indices[:num_features_to_display]]}")
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

# Predict probabilities for the test set
y_pred_prob = gbc.predict_proba(X_test_new)[:, 1]

# Compute the ROC AUC score
roc_auc = roc_auc_score(y_test, y_pred_prob)
print('ROC AUC score:', roc_auc)

# Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

# Plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
